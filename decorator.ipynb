Decorators are a powerful and flexible features in python that allows you to modify the behaviour of a function or class method. They are commonly used to add functionality to functions or methods without modifying their actual code.

```python
#function copy
#closures
#decorators
```

```python
#function copy
def welcome():
    return "Welcome to HCL"

welcome()
```

```python
wel=welcome
print(wel())
del welcome
print(wel())
```

```python
#closure
def main_welcome():
    msg="Welcome"
    def sub_welcome_method():
        print("Welcome to HCl")
        print(msg)
        print("Please behave according to human")
    return sub_welcome_method()
```

```python
main_welcome()
```

```python
def main_welcome(func):
    
    def sub_welcome_method():
        print("Welcome to the course")
        func()
        print("Welcome to HCL")
    return sub_welcome_method()
```

```python
main_welcome(print)
```
