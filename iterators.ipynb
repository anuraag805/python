Iterators are advanced Python Concepts that allow for efficient looping and memory management. Iterators provide a way to access elements of a collection sequentially without exposing the underlyig structure.

```python
my_list=[1,2,3,4,5,6]
for i in my_list:
    print(i)
```

```python
type(my_list)
```

```python
#Iterator
iterator=iter(my_list)
print(type(iterator))
```

```python
iterator
```

```python
#Iterator through all the element
next(iterator)
```

```python
iterator=iter(my_list)
```

```python


try:
    print(next(iterator))
except StopIteration:
    print("Stop execution after the list is over")
```
