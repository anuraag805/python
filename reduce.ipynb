```python
from functools import reduce

numbers = [1, 2, 3, 4, 5]

result = reduce(lambda x, y: x + y, numbers)

print(result)
```

```python
from functools import reduce

def add(x, y):
    return x + y

numbers = [10, 20, 30]

result = reduce(add, numbers)

print(result)
```

```python
from functools import reduce

numbers = [1, 2, 3]

result = reduce(lambda x, y: x + y, numbers, 10)

print(result)
```

Reduce needed when you need single output value
